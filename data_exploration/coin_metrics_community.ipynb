```python
import requests
import pandas as pd

url = "https://community-api.coinmetrics.io/v4/timeseries/asset-metrics"
params = {
    "assets": "btc",
    "metrics": "PriceUSD,TxCnt,AdrActCnt",  # all free metrics
    "frequency": "1d",
    "start_time": "2024-01-01",
    "end_time": "2024-12-31"
}

response = requests.get(url, params=params)
print("Status code:", response.status_code)
```

```python
data = response.json()

df = pd.DataFrame(data["data"])
df["time"] = pd.to_datetime(df["time"])
df[["PriceUSD", "TxCnt", "AdrActCnt"]] = df[["PriceUSD", "TxCnt", "AdrActCnt"]].astype(float)
```
